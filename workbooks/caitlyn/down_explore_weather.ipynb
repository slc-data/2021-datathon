```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm

import wrangle
```

```python
weather = wrangle.wrangle_weather()
```

```python
weather['wind'] = weather['wind'].str.extract('(\d+)', expand=False)
weather['visibility'] = weather['visibility'].str.extract('(\d+)', expand=False)
```

```python
weather["wind"].fillna(0, inplace = True)
```

```python
weather['wind'] = weather['wind'].astype(int)
weather['visibility'] = weather['visibility'].astype(int)
```

```python
weather_train, weather_validate, weather_test = wrangle.split_data(weather)
```

# Univariate

```python
plt.figure(figsize=(30,5))
sns.countplot(data=weather, x='celsius')
plt.title('Count of Celcius Readings')
plt.xlabel('Temp in Celcius')
plt.ylabel('Count of Times Sensed')
plt.show()
```
